三. 使用BS4套件進行解析，取得所需資料

3.1. 使用parser作為解析器

```python
soup = BeautifulSoup(res.text, "html.parser")
```

3.2. 尋找資料所在的標籤

```python
soup.find_all("a")
```

3.3. 往資料所在標籤的上一層查詢，篩除無用資料

```python
soup.find_all("span", {"class":"thread-title"})
```

3.4.結合兩次的find_all查詢，並只保留文字內容

```python

result = soup.find_all("span", {"class":"thread-title"})

for r in result:
  tag_a = r.find_all("a")[0]
  print(tag_a.get_text())
```

3.5.練習修改成爬取留言

```python
result = soup.find_all("span", {"class":"thread-popularity"})
for r in result:
  print(r.get_text())
```

四. 整合上方程式碼，一次爬取多頁的內容。

https://pttweb.tw/news?page=1, https://pttweb.tw/news?page=2, https://pttweb.tw/news?page=3

4.1.爬取標題內容

```python
from bs4 import BeautifulSoup
import requests

for p in range(1,3):
  res = requests.get('https://pttweb.tw/news?page=%s' % p)
  soup = BeautifulSoup(res.text, "html.parser")
  result = soup.find_all("span", {"class":"thread-title"})
  for r in result:
    tag_a = r.find_all("a")[0]
    print(tag_a.get_text())
```

4.1.修改爬取日期，並放到list中，再print出資料

```python
from bs4 import BeautifulSoup
import requests

data = []
for p in range(1,3):
  res = requests.get('https://pttweb.tw/news?page=%s' % p)
  soup = BeautifulSoup(res.text, "html.parser")
  result = soup.find_all("span", {"class":"thread-posttime"})
  for r in result:
    data.append(r.get_text())
print(data)
```
